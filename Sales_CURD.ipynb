```python
import pandas as pd
import os

# File path to the CSV file
file_path = 'sales_data.csv'
encoding = 'ISO-8859-1'

#Craete new record
def create_record(new_record):
    try:
        if not os.path.isfile(file_path):
            print("File does not exist. Please check the file path.")
            return
        
        # Load the existing dataset
        df = pd.read_csv(file_path, encoding=encoding)
        
        # Convert new record to DataFrame
        new_record_df = pd.DataFrame([new_record])
        
        # Append the new record
        df = pd.concat([df, new_record_df], ignore_index=True)
        
        # Save the updated dataset back to the CSV file
        df.to_csv(file_path, index=False, encoding=encoding)
        print("Record added successfully.")
    except PermissionError as e:
        print(f"Permission Error: {e}")
    except Exception as e:
        print(f"Error: {e}")

# Read all records
def read_records(condition=None):
    try:
        if not os.path.isfile(file_path):
            print("File does not exist. Please check the file path.")
            return
        
        # Load the dataset
        df = pd.read_csv(file_path, encoding=encoding)
        
        # If a condition is provided, filter the records
        if condition:
            df = df.query(condition)
        
        print("\nRecords:")
        print(df)
    except PermissionError as e:
        print(f"Permission Error: {e}")
    except Exception as e:
        print(f"Error: {e}")

#update the records
def update_record(index, updates):
    try:
        if not os.path.isfile(file_path):
            print("File does not exist. Please check the file path.")
            return
        
        df = pd.read_csv(file_path, encoding=encoding)

        for key, value in updates.items():
            df.at[index, key] = value
        
        df.to_csv(file_path, index=False, encoding=encoding)
        print("Record updated successfully.")
    except PermissionError as e:
        print(f"Permission Error: {e}")
    except Exception as e:
        print(f"Error: {e}")

#Delete a record
def delete_record(index):
    try:
        if not os.path.isfile(file_path):
            print("File does not exist. Please check the file path.")
            return
        
        df = pd.read_csv(file_path, encoding=encoding)
        
        df = df.drop(index)
        
        df.reset_index(drop=True, inplace=True)
        
        df.to_csv(file_path, index=False, encoding=encoding)
        print("Record deleted successfully.")
    except PermissionError as e:
        print(f"Permission Error: {e}")
    except Exception as e:
        print(f"Error: {e}")

# Example usage

# Create a new record
new_record = {
    'ORDERNUMBER': 4,           # Assuming 'ORDERNUMBER' is a unique identifier
    'PRODUCTLINE': 'Motorcycles',
    'QUANTITYORDERED': 10,
    'PRICEEACH': 19.99
}
create_record(new_record)
create_record(new_record)

# Read records 
read_records() 
read_records("QUANTITYORDERED > 95") 

# Update a record 
updates = {
    'PRICEEACH': 25.99
}
update_record(index=2, updates=updates)

# Delete a record
delete_record(index=1)
```
